The analysis posted below is a baseline analysis of the Titanic Data Set using simple functions in numpy and pandas.

```python
import numpy as np
import pandas as pd
titanic=pd.read_csv("train.csv")
titanic
```

```python
titanic.isnull().sum()
```

```python
len (titanic.index)
```

```python
df2=titanic.drop(columns=['Cabin'])
df2
df2=df2.dropna(subset=['Embarked'])
df2
```

```python
age_analysis = df2.groupby('SibSp')['Age'].agg(['count', 'mean']).reset_index()
age_analysis
```

```python
df2.loc[df2.Age.isnull(),'Age'] = df2.groupby('SibSp').Age.transform('mean').round(1)
df2
```

```python
bins = [0,10,20,30,40,50,60,90]
bin_labels = ['<10','10-19','20-29', '30-39', '40-49', '50-59', '>60']
df2['AgeGr'] = pd.cut(df2.Age, bins, right=False, labels = bin_labels)
```

```python
age_analysis = df2.groupby('AgeGr')['Parch'].agg(['count', 'mean','sum']).reset_index()
age_analysis
```

I engineered the attribute, AgeGr (Age Group). I did this because the age category was not concise and I felt that the Age was an important attribute to include. So, I created Age Group, which assigns each Age into a category (<10, 10-19, 20-29, 30-39, 40-49, 50-59, >60) in order to provide a snapshot into the survival rate of each age group. I used AgeGr as an alternative to Age.

```python
sexage_Analysis = pd.crosstab(df2['Sex'],df2['AgeGr'],df2['Survived'],aggfunc='mean').round(1)
sexage_Analysis
```

```python
sexclass_Analysis = pd.crosstab(df2['Sex'],df2['Pclass'],df2['Survived'],aggfunc='mean')
sexclass_Analysis
```

```python
ageclass_Analysis = pd.crosstab(df2['AgeGr'],df2['Pclass'],df2['Survived'],aggfunc='mean')
ageclass_Analysis
```

```python
ageparch_Analysis = pd.crosstab(df2['AgeGr'],df2['Parch'],df2['Survived'],aggfunc='mean')
ageparch_Analysis
```

```python
Parchclass_Analysis = pd.crosstab(df2['Parch'],df2['Pclass'],df2['Survived'],aggfunc='mean')
Parchclass_Analysis
```

```python
Sexparch_Analysis = pd.crosstab(df2['Sex'],df2['Parch'],df2['Survived'],aggfunc='mean')
Sexparch_Analysis
```

According to my analysis, the most likely group to survive are younger females, who have a small family and sat in first class. For my aggregate statistic, I used the mean because the data was binary and with binary data, you get the chance of survival, which is very useful for such an analysis. We can see this through the many dataframes. As for analysis in regards to Sex and Class, where a woman in first class has a 96.7% chance of survival, whereas a male in the same class has a 36.9% chance of survival. This observation is validated by looking at the other classes and realizing that in general, women are more likely to survive and the higher the class, the higher the survival rate would be. As for Age and Sex, women had a very high chance of survival, have a survival rate between 70%-100%, whereas men older than 10 years, had a low chance of survival, having a survival rate between 10%-20%. The Age and Class Analysis further validate the findings stated that the higher the class, the higher the chance at survival. As for families, I used the parent/child statistic because it had more representation throughout the earlier stated categories. In this, I found that the smaller the family, the higher the chance at survival. This can be seen in all tables above with pertinence to Parch.

I excluded PassengerID, Name, Ticket, Fare, Cabin and Embarked. I excluded PassengerID because there was no real correlation in this category, so I felt that the information gathered from this category would not provide pertinent information. I excluded Name and Ticket for a similar reason as PassengerID, as it did not provide pertinent information for the analysis needed. Fare was not included because it provided similar insight as Pclass (the more one paid for a ticket meant the higher the class they sat) but the values were a bit less concise. I did not include Cabin because this category had around 77% of the data missing, which meant that the analysis of this data would not lead to a good conclusion about the Titanic in general. Finally, I left out Embarked because it did not appear to heavily influence the survival rate (ie: where you embarked did not have as much of a say in whether or not someone survived in my opinion).

As for the missing values in the data, I treated each category different, as each category represented a different scenario. The three categories with missing data were Cabin, Embarked and Age. For Cabin, I realized that 77% of the data was missing. Since there was so much information missing, it would be really difficult for me to predict the missing values, so I felt that it did not have much pertinence for my data analysis, so I deleted the column. For the missing Embarked values, since it was less than 1%, 2/891 passengers, I deleted those entries. For Age, I took a different approach. I realized that around 15% of the data was missing. So, I looked at the other completed attribute to see which one predicted age the best. I realized the SibSp was a good indicator since the mean age for each section of SibSp accurately predicted the Age of a prospective person in such a scenario. For example, if someone were travelling alone or with one Sibling/Spouse, they were probably a person in their 20s or 30s, which was accurately predicted by the SibSp attribute. I did recognize that there would be some degree of error, but since Age was such a vital attribute in my prediction model, I chose to use this prediction in hope that the error would be minimized and that the chance of survival in pertinence to Age would be known.
